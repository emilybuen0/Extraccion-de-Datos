**b) Host unidos después del 2020**

```python
#Reemplacé la / por -
#Convertí de cadena a datetime
df['host_since'] = df['host_since'].str.replace('/','-')
df['host_since'] = pd.to_datetime(df['host_since'], format = "%d-%m-%Y",errors='coerce')
dsps = df[df['host_since'] > pd.Timestamp("2020-01-01")]
dsps
```

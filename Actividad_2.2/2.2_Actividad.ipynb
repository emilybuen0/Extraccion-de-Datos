```python
import json
import glob
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("HongKong_China_Data.csv")
df
```

**a) “host_response_rate” sea mayor a 95% y que no sean “Superhost”**

```python
hrrsh = df[(df["host_response_rate"] >= '95%') & (df["host_is_superhost"] == 'f')]
hrrsh
```

**b) Los registros cuyas propiedades no tengan “1 baño”**

```python
banos = df[(df["bathrooms"] > 1)] 
banos
```

**c) Los registros que tengan un precio entre “$100 y $200 US” (Realizar conversión de moneda)**

```python
df['host_since'] = df['host_since'].str.replace('$','')
df['host_since'] = df['host_since'].str.replace(',','')
precio = df[(df["price"] <= "200.00") | (df["price"] >= "100.00")]
precio
```

**d) Los registros que sean “instant_bookable” y que hallan verificado su perfil “host_identity_verified”**

```python
ibiv = df[(df["instant_bookable"] == 't') & (df["host_identity_verified"] == 't')]
ibiv
```

**e) Los registros de los tipos de cuarto: “hotel room” o “Private room”**

```python
hrpr = df[(df["room_type"] == 'Hotel room') | (df["room_type"] == 'Private room')]
hrpr
```

**f) Los registros que tengan mas de 4 “accommodates”**

```python
accommodates = df[(df["accommodates"] > 4)]
accommodates
```

```python
hrrsh.to_csv("A")
banos.to_csv("B")
precio.to_csv("C")
ibiv.to_csv("D")
hrpr.to_csv("E")
accommodates.to_csv("F")
```
